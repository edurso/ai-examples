Dataset from [HackerRank 'Stack Exchange Question Classifier' Challenge](https://www.hackerrank.com/challenges/stack-exchange-question-classifier/problem)

```python
import json
import re
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import BernoulliNB
```

```python
i = 0
x = []
y = []
```

```python
trainingfile = open('training.json', 'r')
```

```python
for json_ in trainingfile:
    i += 1
    if i == 1: 
        continue
    post = json.loads(json_)
    y.append(post['topic'])
    x.append(" ".join(re.findall(r'\w+', post['question'] + '\n\n' + post['excerpt'])))
```

```python
trainingfile.close()
```

```python
X = np.array(x) 
Y = np.array(y)
```

```python
model = Pipeline([
    ('vectorizer',  CountVectorizer()),
    ('tfidf_transformer',  TfidfTransformer()),
    ('classifier',  MultinomialNB()) 
])
model.fit(X,Y)
```

```python
tx = []
i = 0
testfile = open('testcases.json', 'r')

for json_ in testfile:
    i += 1
    if i == 1: 
        continue
    post = json.loads(json_)
    tx.append(" ".join(re.findall(r'\w+', post['question'] + '\n\n' + post['excerpt'])))
    
testfile.close()
    
tX = np.array(tx)

yhats = model.predict(tX)
for yhat in yhats:
    print(yhat)
```
